```python
remove=['Poster_Link', 'Certificate', 'Meta_score', 'Star1', 'Star2', 'Star3', 'Star4', 'No_of_Votes', 'Gross']
data.drop(remove,axis=1,inplace=True)
```

```python
genres=data['Genre'].tolist()
movies=data['Series_Title'].tolist()
```

```python
final={}
for x in range(1000):
    final[movies[x]]=list(genres[x].split(', '))
```

```python
final
```

```python
drama=[]
action=[]
romance=[]
comedy=[]
crime=[]
adventure=[]
western=[]
sci_fi=[]
biography=[]
fantasy=[]
history=[]
thriller=[]
animation=[]
family=[]
mystery=[]
horror=[]
```

```python
for i in final:
    if "Drama" in final[i]:
        drama.append(i)
    if "Romance" in final[i]:
        romance.append(i)
    if "Action" in final[i]:
        action.append(i)
    if "Comedy" in final[i]:
        comedy.append(i)
    if "Crime" in final[i]:
        crime.append(i)
    if "Adventure" in final[i]:
        adventure.append(i)
    if "Western" in final[i]:
        western.append(i)
    if "Sci-Fi" in final[i]:
        sci_fi.append(i)
    if "Biography" in final[i]:
        biography.append(i)
    if "Fantasy" in final[i]:
        fantasy.append(i)
    if "History" in final[i]:
        history.append(i)
    if "Thriller" in final[i]:
        thriller.append(i)
    if "Animation" in final[i]:
        animation.append(i)
    if "Family" in final[i]:
        family.append(i)
    if "Mystery" in final[i]:
        mystery.append(i)
    if "Horror" in final[i]:
        horror.append(i)
```
